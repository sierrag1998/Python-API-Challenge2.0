```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import gmaps
import os
import json

# Import API key
from api_keys import g_key
```

```python
#Original Data 

# Output File (CSV)
file = "../starter_code/cities_df.csv"
cities_df = pd.read_csv(file)
cities_df.head()
```

```python
cities_df.shape
```

```python
clean_df = cities_df.dropna(how="any")
```

```python
#Humidity Heatmap 

# Configure gmaps with API key
gmaps.configure(api_key=g_key)

# Store 'Lat' and 'Lng' into  locations 
locations = clean_df[["Lat", "Long"]]
humidities = clean_df["Humidity"].astype(float)

# Create a Heatmap layer
fig = gmaps.figure(zoom_level=1, center=[0,0])

heat_layer = gmaps.heatmap_layer(locations, weights= humidities, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 3)

fig.add_layer(heat_layer)

fig
```

```python
#Ideal Cities 



ideal_df = clean_df.loc[(clean_df["Max Temp"] > 70) & (clean_df["Max Temp"] < 80) & (clean_df["Cloudiness"] == 0) & (clean_df["Wind Speed"] < 10),:] 
 
ideal_final = ideal_df.drop([ideal_df.index[6]])
                             
ideal_final
                      
```

```python
# Find Hotels - Lagoa
 
target_coordinates = "39.05, -27.98"
target_search = "Hotel"
target_radius = 5000
target_type = "Hotel"

# set up a parameters dictionary
params = {
    "location": target_coordinates,
    "keyword": target_search,
    "radius": target_radius,
    "type": target_type,
    "key": g_key
}

# base url
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"

# run a request using our params dictionary
response = requests.get(base_url, params=params)

# convert response to json
places_data = response.json()

# Print the name and address of the first restaurant that appears
print(places_data["results"][0]["name"])
print(places_data["results"][0]["vicinity"])
```

```python
#Add Hotels 


#Create Hotel DF
hotel_list = ({ "City" : ["lagoa", "saint-denis", "erzin", "corinto", "mattoon", "mao"],
    "Hotel": ["Graciosa Resort","ibis Saint-Denis Stade Ouest","Hattusa Vacation Thermal Club Erzin","Club Hotel Casino Loutraki","Comfort Suites","ARTIEM Capri"]})
hotel_list_df = pd.DataFrame(hotel_list)


#Join

merge = pd.merge(ideal_final,hotel_list_df, on="City")

merge
```

```python
#Add Hotel Markers to Heat Map

# Convert Hotels as a list

hotels = merge["Hotel"].tolist()
cities = merge["City"].tolist()
countries = merge["Country"].tolist()

info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""

# Create layered map 

fig = gmaps.figure()

locations2 = merge[["Lat", "Long"]]

heat_layer = gmaps.heatmap_layer(locations, weights= humidities, 
                                 dissipating=False, max_intensity=100,
                                 point_radius = 3)

hotel_info = [info_box_template.format(**row) for index, row in merge.iterrows()]




markers = gmaps.marker_layer(locations2,
    info_box_content=hotel_info)
fig.add_layer(heat_layer)
fig.add_layer(markers)
fig
```
